```python
#https://medium.com/@randerson112358/stock-price-prediction-using-python-machine-learning-e82a039ac2bb
import math
from matplotlib import style
import datetime as dt
import pandas_datareader as web
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, RobustScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import ta

plt.style.use('ggplot')

start = dt.datetime(2000, 1, 1)
end = dt.datetime.now()
df = web.DataReader("0005.HK", 'yahoo', start, end)
#df.reset_index(inplace=True)
#df.set_index("Date", inplace=True)
#df = df.drop("Symbol", axis=1)
#df.to_csv('TSLA.csv')
#df = pd.read_csv('tsla.csv', parse_dates=True, index_col=0)
```

```python
#Visualize the closing price history
plt.figure(figsize=(16,8))
plt.title('Close Price History')
plt.plot(df['Close'])
plt.xlabel('Date',fontsize=18)
plt.ylabel('Close Price USD ($)',fontsize=18)
plt.show()
```

```python
# Adding all the indicators
df = ta.add_all_ta_features(df, open="Open", high="High", low="Low", close="Close", volume="Volume", fillna=True)
#Converting the dataframe to a numpy array
dataset = df.values
#print(dataset)
#Create a new dataframe with only the 'Close' column
close_data = df.filter(['Close'])
#Get /Compute the number of rows to train the model on
training_data_len = math.ceil( len(dataset) *.8) 
```

```python
#Scale the all of the data to be values between 0 and 1 
#scaler = MinMaxScaler(feature_range=(0, 1)) 
scaler = RobustScaler()
scaled_data = scaler.fit_transform(close_data)

#scaler_for_training = MinMaxScaler(feature_range=(0, 1)) 
scaler_for_training = RobustScaler()
dataset = scaler_for_training.fit_transform(dataset)
```

```python
#Create the scaled training data set 
train_data = dataset[0:training_data_len  , : ]
#Split the data into x_train and y_train data sets
x_train = []
y_train = []
look_forward_days=7
look_backward_days=30
for i in range(look_backward_days,len(train_data)-look_forward_days+1):
    x_train.append(train_data[i-look_backward_days:i])
    y_train.append(list(train_data[i:i+look_forward_days,0]))
```

```python
#Convert x_train and y_train to numpy arrays
x_train = np.array(x_train)
y_train = np.array(y_train)
#Reshape the data into the shape accepted by the LSTM
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],x_train.shape[2]))
```

```python
#Build the LSTM network model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True,input_shape=(x_train.shape[1],x_train.shape[2])))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=25))
model.add(Dense(units=look_forward_days))
#Compile the model
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
```

```python
#Train the model
model.fit(x_train, y_train, batch_size=1, epochs=5, validation_split=0.5)
```

```python
#Test data set
test_data = dataset[training_data_len - look_backward_days: , : ]
#Create the x_test and y_test data sets
x_test = []
y_test = []
for i in range(look_backward_days,len(test_data)):
    x_test.append(test_data[i-look_backward_days:i])
    y_test.append(list(train_data[i:i+look_forward_days,0]))
y_test = np.array(y_test)
print(y_test.shape)
print(len(train_data))
print(len(x_train))
print(len(x_test))
```

```python
#Convert x_test to a numpy array 
x_test = np.array(x_test)
#Reshape the data into the shape accepted by the LSTM
x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1],x_test.shape[2]))
#Getting the models predicted price values
predictions = model.predict(x_test) 
print(y_test.shape)
predictions = scaler.inverse_transform(predictions)#Undo scaling
#Calculate/Get the value of RMSE
rmse=np.sqrt(np.mean(((predictions- y_test)**2)))
rmse
```

```python
#Plot/Create the data for the graph
train = close_data[:training_data_len]
valid = close_data[training_data_len:]
new_predictions = predictions.T
#print(predictions)

length = len(new_predictions)-5
prediction_dict=[]
for i in range(length):
    prediction_dict.append("Predictions"+str(i))
for i in range(length):
    valid["Predictions"+str(i)] = new_predictions[i]

#Visualize the data
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.plot(train['Close'])
plt.plot(valid['Close'])
plt.plot(valid[prediction_dict])
x=['Train', 'Val']
for i in range(length):
    x.append(prediction_dict[i])
plt.legend(x, loc='lower right')
#plt.legend(['Train', 'Val', 'Predictions0'], loc='lower right')
plt.show()
```
